## Setting Random Seeds for Reproducibility

To ensure that the results of this analysis are consistent and reproducible, I define a function to set random seeds across various libraries. This practice is crucial when working with stochastic processes in machine learning and deep learning models.

### Function Overview

The `set_random_seeds()` function is designed to initialize random seeds across the `numpy`, `random`, and `TensorFlow` libraries. It also configures the verbosity of the XGBoost library to minimize output during training.

### Workflow

1. **Setting Seed for Numpy**:
   - The `np.random.seed()` function is used to set the seed for `numpy`'s random number generator. This step ensures that any random operations in `numpy` produce the same results across different runs.

2. **Setting Seed for Python's Random Module**:
   - The `random.seed()` function sets the seed for Python's built-in `random` module. This ensures that functions like `random.random()` and `random.shuffle()` generate consistent results.

3. **Setting Seed for TensorFlow**:
   - TensorFlow's random number generator is seeded using `tf.random.set_seed()`. This is particularly important when training neural networks, as it ensures that the initial weights and the order of batch processing are the same in every run.

4. **Configuring XGBoost**:
   - The verbosity of the XGBoost library is set to 0 to suppress unnecessary output during model training. Additionally, the seed for XGBoost's random processes is set using the `xgparam` dictionary.

### Conclusion

By setting random seeds across different libraries, I ensure that the results are reproducible. This practice is a cornerstone of reliable and rigorous data science, enabling others to replicate and validate the findings.


```python
# Function to set random seeds for reproducibility
def set_random_seeds(seed_value=42):
    """
    Configure random seed for reproducibility.
    Args:
    - seed_value: Seed for RNGs (default: 42).
    Returns:
    - None
    """
    # Set seed for numpy RNG
    np.random.seed(seed_value)

    # Set seed for Python's random RNG
    random.seed(seed_value)

    # Set seed for TensorFlow RNG
    tf.random.set_seed(seed_value)

    # Configure XGBoost verbosity to 0 (silent)
    xgb.set_config(verbosity=0)

    # Set seed for XGBoost parameters
    xgparam = {'seed': seed_value}

# Initialise random seed for consistency
set_random_seeds()
```

## Downloading and Processing Stock Data

This section outlines the process of downloading and processing stock market data for multiple tickers over specified time periods. The code is structured into several functions that handle different aspects of data retrieval and preparation.

### Function Overview

The code includes multiple functions:

1. **`fetch_data(ticker, start, end, output_dir)`**:
   - Downloads the stock data for a specific ticker from Yahoo Finance and saves it as a CSV file in the specified directory. If the data already exists, it simply loads it from the file.

2. **`process_data(df, ticker)`**:
   - Processes the raw data by extracting the 'Adj Close' prices and renaming this column to match the ticker symbol. This function ensures that the data is consistent and ready for analysis.

3. **`download_and_process_data(tickers, start, end, out_dir)`**:
   - Handles the download and processing of data for multiple tickers. The function combines the processed data into a single DataFrame, making it easier to manage and analyze.

4. **`main()`**:
   - The central function that orchestrates the downloading and processing of stock data for different datasets (training, testing, and extra testing). It specifies the tickers and date ranges for each dataset, calls the relevant functions, and prints out the first few rows of each combined dataset.

### Workflow

1. **Fetching and Saving Data**:
   - The `fetch_data()` function is responsible for downloading the data for a given ticker from Yahoo Finance within the specified date range. If the data already exists in the specified directory, it loads the data from the existing CSV file. This function ensures that all stock data is stored locally, reducing the need for repeated downloads.

2. **Processing the Data**:
   - Once the data is fetched, the `process_data()` function extracts the 'Adj Close' column, which represents the adjusted closing price. This column is renamed to the ticker symbol for easier identification. The processed data is then ready for merging with other tickers.

3. **Combining Data for Multiple Tickers**:
   - The `download_and_process_data()` function loops through the list of tickers, downloading and processing the data for each one. The processed data for each ticker is then merged into a single DataFrame, ensuring that all data is aligned by date.

4. **Managing Data Download and Processing**:
   - The `main()` function defines the tickers for training, testing, and extra testing datasets. It specifies the date ranges for each dataset and directories where the data should be saved. The function calls `download_and_process_data()` for each set of tickers, resulting in three combined datasets. It then prints the first few rows of each dataset to verify the download and processing steps.

### Conclusion

By following this structured approach, I efficiently download, process, and combine stock data for multiple tickers across different time periods. This setup ensures that all data is consistently prepared and ready for further analysis, such as training machine learning models. The use of functions for each step of the process makes the code modular, easier to maintain, and reusable for different datasets.


```python
def fetch_data(ticker, start, end, output_dir):
    """
    Fetch and save stock data for a ticker.
    Args:
    - ticker (str): Stock ticker symbol.
    - start (str): Start date (YYYY-MM-DD).
    - end (str): End date (YYYY-MM-DD).
    - output_dir (str): Directory to save data.
    Returns:
    - pd.DataFrame: Loaded data for ticker.
    """
    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Define file path for the data
    file_path = os.path.join(output_dir,
                             f"{ticker}.csv")

    # Download data
    if not os.path.exists(file_path):
        data = yf.download(ticker, start=start, end=end)
        data.to_csv(file_path)

    # Load and return the data as DataFrame
    return pd.read_csv(file_path)


def process_data(df, ticker):
    """
    Process data: keep 'Adj Close', rename.
    Args:
    - df (pd.DataFrame): Stock data.
    - ticker (str): Stock ticker symbol.
    Returns:
    - pd.DataFrame: Processed data with 'Adj Close'.
    """
    # Convert 'Date' to datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Keep only 'Date' and 'Adj Close' columns
    df = df[['Date', 'Adj Close']]

    # Rename 'Adj Close' to ticker symbol
    df = df.rename(columns={'Adj Close': ticker})

    # Return the processed DataFrame
    return df


# Function to download and process data
def download_and_process_data(tickers, start, end, out_dir):
    """
    Download and process stock data.
    Args:
    - tickers (list): List of stock tickers.
    - start (str): Start date (YYYY-MM-DD).
    - end (str): End date (YYYY-MM-DD).
    - out_dir (str): Directory to save data.
    Returns:
    - pd.DataFrame: Combined data for tickers.
    """
    # Initialize empty DataFrame
    combined_df = None

    # Loop through each ticker
    for ticker in tickers:
        # Fetch raw data
        raw_df = fetch_data(ticker, start, end, out_dir)

        # Process data for current ticker
        processed_df = process_data(raw_df,
                                    ticker.split('.')[0])

        # Merge processed data with combined_df
        combined_df = (processed_df if combined_df is None
                       else pd.merge(combined_df, processed_df,
                                     on='Date', how='inner'))

    # Return the combined DataFrame
    return combined_df


    # Define tickers for each dataset
    tickers_train = ['SVT.L', 'CCH.L', 'SHEL.L',
                     'ANTO.L', 'LGEN.L']
    tickers_test = ['RR.L', 'SGE.L','LAND.L',
                    'SN.L', 'AAL.L','BDEV.L', 'WPP.L',
                    'UU.L', 'GSK.L', 'HSBA.L']
    tickers_extra =['ULVR.L', 'NG.L', 'BARC.L',
                     'TSCO.L', 'SBRY.L',
                     'BP.L',  'AZN.L',
                    'IMB.L', 'SPX.L', 'RTO.L']

    # Define date ranges for datasets
    date_train = ('2014-01-01', '2023-12-31')
    date_test = ('2024-01-01', '2024-03-31')
    date_extra = ('2024-01-01', '2024-06-30')

    # Directories for saving data
    dirs = ('train_data', 'test_data', 'extra_data')

    # Download and process training data
    combined_train_data = download_and_process_data(
        tickers_train, *date_train, dirs[0])

    # Download and process test data
    combined_test_data = download_and_process_data(
        tickers_test, *date_test, dirs[1])

    # Download and process extra data
    combined_extra_data = download_and_process_data(
        tickers_extra, *date_extra, dirs[2])

    # Print first few rows of each dataset
    print(combined_train_data.head())
    print(combined_test_data.head())
    print(combined_extra_data.head())

# Execute main function when script is run
if __name__ == "__main__":
    main()
```

## Exploratory Data Analysis (EDA) on Stock Data

In this section, I perform Exploratory Data Analysis (EDA) on the stock data to understand its structure, identify patterns, and detect potential anomalies. EDA is a crucial step in the data preprocessing pipeline as it provides insights that guide subsequent modeling and analysis.

### Function Overview

The central function in this process is `perform_eda`, which handles the exploratory analysis of the stock data. It generates key statistical summaries, visualizations, and logs the findings. The `main` function orchestrates the entire process, from downloading and processing the data to performing EDA on different datasets (train, test, and extra).

### Workflow

1. **Fetching and Processing Data**:
   - Stock data for multiple tickers is fetched and processed using the `download_and_process_data` function. This function retrieves historical stock prices and processes them into a format suitable for analysis and modeling.

2. **Performing EDA**:
   - The `perform_eda` function is applied to the processed datasets. It performs the following key tasks:
     - **Basic Statistical Analysis**: Descriptive statistics are computed for each stock ticker, including metrics such as mean, standard deviation, and quartiles. These statistics are saved as Excel files for each ticker.
     - **Missing Values Analysis**: The function checks for and reports any missing values in the datasets, which is crucial for ensuring data quality.
     - **Correlation Analysis**: A correlation matrix is computed and visualized as a heatmap. This matrix shows the relationships between the different stock tickers, helping to identify whether certain stocks move together.
     - **Log Returns Calculation**: Log returns are calculated for each ticker, providing insights into the daily percentage change of stock prices. The log returns are plotted over time to visualize the volatility and trends of each stock.

3. **Saving and Displaying Results**:
   - The results of the EDA, including statistical summaries, correlation heatmaps, and log returns plots, are saved to the specified directories. These outputs are essential for documenting the findings and preparing the data for further analysis.

### Conclusion

By performing EDA on the stock data, I gain a comprehensive understanding of the dataset, including its statistical properties, correlations, and price movements. The insights gathered from this analysis are critical for guiding the development of predictive models and for ensuring that the data is well-prepared and reliable. The saved plots and statistics also provide a valuable reference for future analysis and decision-making.


```python
def perform_eda(df, data_type, output_dir):
    """
    Perform Exploratory Data Analysis (EDA) on the provided DataFrame.

    Args:
    - df (pd.DataFrame): The data containing stock prices for multiple tickers.
    - data_type (str): Indicates whether the data is 'Train', 'Test', or 'Extra'.
    - output_dir (str): Directory where the results (plots and statistics) will be saved.

    Returns:
    - None: This function saves the plots and statistics in the specified directory.
    """
    # Ensure that the output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Display and save basic statistical information
    print(f"\nBasic Stats ({data_type}):")
    basic_stats = df.describe()  # Generate descriptive statistics
    pd.options.display.float_format = "{:.2f}".format  # Format floats to 2dp
    print(basic_stats)  # Display the statistics

    # Save basic statistics to Excel files, one per column
    for column in basic_stats.columns:
        column_stats = basic_stats[[column]] # Select stats for column
        column_stats.to_excel(
            os.path.join(output_dir,
                         f'stats_{data_type.lower()}_{column}.xlsx')
        )# save to Excel

    # Check and display any missing values in the data
    print(f"\nMissing Values ({data_type}):")
    missing_values = df.isnull().sum() # Count missing values per column
    print(missing_values) # Display missing values

    # Calculate and display the correlation matrix
    print(f"\nCorrelation Matrix ({data_type}):")

    # Compute correlation matrix without date
    corr_matrix = df.drop('Date', axis=1).corr()

    # Display the correlation matrix
    print(corr_matrix)

    # Plot a heatmap to visualize the correlation matrix
    print(f"\nCorrelation Heatmap ({data_type}):")

    # Set figure size for heatmap
    plt.figure(figsize=(12, 8))

    # Plot heatmap
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

    # Add title to heatmap
    plt.title(f'Heatmap ({data_type})', fontsize=16)

    # Save heatmap
    plt.savefig(os.path.join(output_dir, f'heatmap_{data_type.lower()}.png'))

    # display Heatmap
    plt.show()

    # Calculate log returns for the stock prices
    log_returns = np.log(df.set_index('Date').pct_change() + 1).dropna()

    # Begin plotting log returns for each stock ticker over time
    print(f"\nLog Returns ({data_type}):")

    # Determine the number of columns (tickers) minus the Date column
    num_columns = len(df.columns) - 1

    # Create subplots for each stock ticker's log returns
    fig, axes = plt.subplots(num_columns, 1, figsize=(14, 3 * num_columns),
                             sharex=False)

    # Adjust spacing between subplots
    fig.subplots_adjust(hspace=0.6)

    # Loop through each stock ticker and plot its log returns
    for i, column in enumerate(log_returns.columns):

        # Plot log returns for the current ticker.
        axes[i].plot(log_returns.index, log_returns[column], label=column)

        # Set the title of the subplot to the ticker name.
        axes[i].set_title(f'{column} Log Returns', fontsize=14)

        # Label the y-axis as 'Log Returns'
        axes[i].set_ylabel('Log Returns', fontsize=12)

         # Add a legend in the upper right corner
        axes[i].legend(loc='upper right')

         # Enable grid lines for clarity.
        axes[i].grid(True)

        # Limit the number of x-axis ticks to 6.
        axes[i].xaxis.set_major_locator(plt.MaxNLocator(6))

         # Rotate x-axis labels by 45 degrees
        axes[i].tick_params(axis='x', rotation=45)

    # Label the x-axis of the last subplot with 'Date'
    axes[-1].set_xlabel('Date', fontsize=12)

    # Tighten layout to prevent overlap of subplots
    plt.tight_layout()

    # Save the log returns plot as a PNG file
    plt.savefig(os.path.join(output_dir,
                             f'log_returns_{data_type.lower()}.png'))
    plt.show()


def main():
    """
    Main function to orchestrate the download, processing, and analysis
     of datasets.
    """
    # Define stock tickers for training data
    tickers_train = ['SVT.L', 'CCH.L', 'SHEL.L', 'ANTO.L', 'LGEN.L']

    # Define stock tickers for testing data
    tickers_test = ['RR.L', 'SGE.L', 'LAND.L', 'SN.L', 'AAL.L',
                    'BDEV.L', 'WPP.L', 'UU.L', 'GSK.L', 'HSBA.L']

    # Define stock tickers for extra testing data
    tickers_extra = ['ULVR.L', 'NG.L', 'BARC.L', 'TSCO.L', 'SBRY.L',
                     'BP.L', 'AZN.L', 'IMB.L', 'SPX.L', 'RTO.L']

    # Set date ranges for training, testing, and extra testing datasets
    date_train = ('2014-01-01', '2023-12-31')
    date_test = ('2024-01-01', '2024-03-31')
    date_extra = ('2024-01-01', '2024-06-30')

    # Specify directories to save the processed datasets
    output_dir_train = 'output/train_data'
    output_dir_test = 'output/test_data'
    output_dir_extra = 'output/extra_data'

    # Download and process the training data
    print("\nDownloading training data...")
    train_data = download_and_process_data(tickers_train, *date_train,
                                           output_dir_train)

    # Download and process the test data
    print("\nDownloading test data...")
    test_data = download_and_process_data(tickers_test, *date_test,
                                          output_dir_test)

    # Download and process the extra test data
    print("\nDownloading extra data...")
    extra_data = download_and_process_data(tickers_extra, *date_extra,
                                           output_dir_extra)

    # Perform EDA on the training dataset
    print("\nPerforming EDA on training data...")
    perform_eda(train_data, 'Train', output_dir_train)

    # Perform EDA on the test dataset
    print("\nPerforming EDA on test data...")
    perform_eda(test_data, 'Test', output_dir_test)

    # Perform EDA on the extra test dataset
    print("\nPerforming EDA on extra data...")
    perform_eda(extra_data, 'Extra', output_dir_extra)

    # Print completion message
    print("\nEDA completed.")
    print(f"Training data saved in {output_dir_train}")
    print(f"Test data saved in {output_dir_test}")
    print(f"Extra data saved in {output_dir_extra}")

    # Save processed data to CSV files
    train_data.to_csv(os.path.join(output_dir_train,
                                   'train_data.csv'), index=False)
    test_data.to_csv(os.path.join(output_dir_test,
                                  'test_data.csv'), index=False)
    extra_data.to_csv(os.path.join(output_dir_extra,
                                   'extra_data.csv'), index=False)

# Execute the main function
if __name__ == "__main__":
    main()
```

## Preprocessing Stock Data for Machine Learning

In this section, I focus on preprocessing stock data, including creating lagged features, scaling the data, and saving the processed data for future use. The goal is to prepare the data so that it is suitable for training machine learning models, particularly for time series forecasting.

### Function Overview

The key functions involved in this process are `preprocess_data` and `main`. The `preprocess_data` function handles the entire preprocessing pipeline, from creating technical indicators and lagged features to scaling and saving the data. The `main` function orchestrates the workflow by loading raw datasets, calling the preprocessing function, and managing the storage of the processed data.

### Workflow

1. **Setting Random Seeds**:

   - I begin by setting random seeds using the `set_random_seeds()` function to ensure reproducibility. This ensures that the results remain consistent across different runs.

2. **Preprocessing Function**:
   - The `preprocess_data` function is responsible for transforming the raw stock data into a format that is suitable for machine learning models. This function includes:

     - **Technical Indicators**: The function computes the 20-day Exponential Moving Average (EMA), 14-day Relative Strength Index (RSI), and Moving Average Convergence Divergence (MACD) for each stock ticker. These indicators help in capturing important trends and momentum in the stock prices.

     - **Lagged Features**: Lagged features are created for each stock ticker. These features represent past values of the stock prices and are crucial for time series forecasting.

     - **Data Scaling**: The features and target variables are scaled using `RobustScaler` and `StandardScaler`, respectively. Scaling is important to standardize the data and improve the performance of machine learning models.
     
     - **Saving Data**: The processed features, target variables, and scalers are saved in the specified output directories. This ensures that the preprocessed data can be easily loaded for model training.

3. **Loading and Preprocessing Data**:
   - The `main` function handles the loading of raw datasets from CSV files, which include training, testing, and extra testing datasets. After loading, it calls the `preprocess_data` function to transform these datasets and save the processed data for future use.

4. **Data Verification**:
   - After preprocessing, the function prints the shapes of the processed datasets to verify that the data has been correctly prepared and that all steps have been executed as intended.

### Conclusion

By following this structured approach, I ensure that the stock data is well-prepared for machine learning tasks, particularly for time series forecasting. The use of technical indicators, lagged features, and data scaling helps in creating a robust dataset that captures the essential characteristics of the stock prices. The saved processed data can be readily used in subsequent modeling steps, making this preprocessing step a crucial part of the machine learning pipeline.


```python
# Set random seeds for reproducibility
set_random_seeds()

def preprocess_data(train_data, test_data, extra_data, n_lags=5,
                    output_dir='processed_data'):
    """
    Preprocess stock data, create lagged features, and save the processed data.
    Args:
    - train_data: Training dataset.
    - test_data: Test dataset.
    - extra_data: Extra test dataset.
    - n_lags: Number of lagged features.
    - output_dir: Directory for saving data.
    Returns:
    - dict: Dictionary with preprocessed data.
    """

    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Initialize dictionary for preprocessed data
    preprocessed_data = {'train': {}, 'test': {}, 'extra': {}}

    # Organize datasets for iteration
    datasets = {'train': train_data, 'test': test_data, 'extra': extra_data}

    # Iterate over each dataset type (train, test, extra)
    for name, data in datasets.items():
        print(f"Processing {name} data")

        # Ensure 'Date' is the index for the data
        if data.index.name != 'Date':
            print(f"Indexing {name} by 'Date'")
            data['Date'] = pd.to_datetime(data['Date'])  # Convert to datetime
            data.set_index('Date', inplace=True)  # Set 'Date' as index

        # Process each stock ticker within the dataset
        for ticker in data.columns:
            print(f"Processing {ticker}")
            stock_data = data[[ticker]].copy()  # Isolate ticker data

            # Compute the 20-day Exponential Moving Average
            stock_data['EMA_20'] = ta.trend.ema_indicator(
                stock_data[ticker], window=20
            )
            # Compute the 14-day Relative Strength Index
            stock_data['RSI'] = ta.momentum.rsi(
                stock_data[ticker], window=14
            )
            # Compute the Moving Average Convergence Divergence
            stock_data['MACD'] = ta.trend.macd(
                stock_data[ticker]
            )

            # Determine how many rows can support lag features
            available_rows = len(stock_data) - n_lags
            if available_rows <= 0:
                print(f"Not enough data for {n_lags} lags.")
                n_lags = len(stock_data) - 1  # Adjust n_lags if necessary

            # Create lagged features for each lag period
            for lag in range(1, n_lags + 1):
                stock_data[f'{ticker}_lag_{lag}'] = stock_data[
                    ticker
                ].shift(lag)  # Shift data to create lag

            # Drop rows with missing values after lagging
            stock_data.dropna(inplace=True)

            # If no data remains, skip this ticker
            if stock_data.empty:
                print(f"No data left for {ticker} in {name}.")
                continue

            # Separate the features (X) from the target (y)
            X = stock_data.drop(columns=[ticker])
            y = stock_data[ticker]

            # Initialize scalers for features and target
            feature_scaler = RobustScaler()
            target_scaler = StandardScaler()

            # Scale the features and target
            X_scaled = feature_scaler.fit_transform(X)
            y_scaled = target_scaler.fit_transform(
                y.values.reshape(-1, 1)
            ).flatten()

            # Convert scaled data back to DataFrame and Series
            X_scaled = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)
            y_scaled = pd.Series(y_scaled, index=y.index)

            # Store the scaled data and scalers in the dictionary
            preprocessed_data[name][ticker] = {
                'X': X_scaled,
                'y': y_scaled,
                'feature_scaler': feature_scaler,
                'target_scaler': target_scaler
            }

            # Save the scaled features to a CSV file
            X_scaled.to_csv(os.path.join(output_dir, f'{name}_{ticker}_X.csv'))

            # Save the scaled target to a CSV file
            y_scaled.to_csv(os.path.join(output_dir, f'{name}_{ticker}_y.csv'))

            # Save the feature scaler for future use
            joblib.dump(feature_scaler, os.path.join(output_dir,
                                        f'{name}_{ticker}_feature_scaler.pkl'))

            # Save the target scaler for future use
            joblib.dump(target_scaler, os.path.join(output_dir,
                                        f'{name}_{ticker}_target_scaler.pkl'))

    # Print the shapes of preprocessed data for verification
    print("\nFinal shapes of preprocessed data:")
    for name, dataset in preprocessed_data.items():
        print(f"\n{name.capitalize()} data:")
        for ticker, data in dataset.items():
            print(f"  {ticker}: X shape {data['X'].shape}, \n"
                  f"y shape {data['y'].shape}")

    # Return the dictionary of preprocessed data
    return preprocessed_data

def main():
    """
    Main function to load datasets, preprocess them, and save the processed data.
    """
    # Define directories for processed data
    output_dir_train = 'output/train_data'
    output_dir_test = 'output/test_data'
    output_dir_extra = 'output/extra_data'

    # Load training data from CSV
    combined_train_data = pd.read_csv(
        os.path.join(output_dir_train, 'train_data.csv')
    )
    # Load test data from CSV
    combined_test_data = pd.read_csv(
        os.path.join(output_dir_test, 'test_data.csv')
    )
    # Load extra test data from CSV
    combined_extra_test_data = pd.read_csv(
        os.path.join(output_dir_extra, 'extra_data.csv')
    )

    # Display the first few rows of each dataset
    print("Train Data:")
    print(combined_train_data.head())
    print("\nTest Data:")
    print(combined_test_data.head())
    print("\nExtra Test Data:")
    print(combined_extra_test_data.head())

    # Preprocess the datasets
    preprocessed_data = preprocess_data(
        combined_train_data, combined_test_data, combined_extra_test_data
    )

# Execute the main function
if __name__ == "__main__":
    main()
```

     - **Appending Features and Targets**:
     
     The function iterates over each stock ticker in the training dataset, extracting the features (`X`) and targets (`y`). These are stored in lists, which are later concatenated.

     - **Concatenation**:
     
     After collecting all the features and targets, the lists are

      `X_train` for features and `y_train` for targets. This unified dataset is more suitable for training machine learning models, which typically require a single input matrix and target vector.

```python
# Set random seeds for reproducibility
set_random_seeds()

def combine_hist_train_datasets(preprocessed_data):
    """
    Combine all individual ticker datasets into a single training dataset.

    Args:
    - preprocessed_data: Dictionary containing preprocessed data for train,
     test, and extra datasets.

    Returns:
    - X_train: Combined features for training.
    - y_train: Combined targets for training.
    """
    combined_X = []
    combined_y = []

    for ticker, data in preprocessed_data['train'].items():
        combined_X.append(data['X'].values)
        combined_y.append(data['y'].values)

    # Concatenate all ticker data into a single dataset
    X_train = np.concatenate(combined_X, axis=0)
    y_train = np.concatenate(combined_y, axis=0)

    return X_train, y_train


```

## Reshaping Data and Feature Verification

In this section, I focus on reshaping the training data to make it suitable for input into a Long Short-Term Memory (LSTM) model and verifying the presence of necessary features across the datasets. This ensures that the data is correctly structured and that no crucial features are missing before proceeding with model training.

### Function Overview

The primary functions used in this process are `reshape_for_lstm` and `check_feature_columns`. The `reshape_for_lstm` function reshapes the input data into sequences of a specified length, which is essential for training LSTM models. The `check_feature_columns` function verifies that all expected features, including lagged features and technical indicators, are present in the datasets.

### Workflow

1. **Random Seed Initialization**:
   - Consistent with best practices, I start by setting random seeds using the `set_random_seeds()` function. This ensures that the results are reproducible across different runs of the code.

2. **Data Reshaping**:
   - The `reshape_for_lstm` function is used to reshape the training data into sequences suitable for LSTM input. The function processes the input data (`X_train`) by creating sequences of a specified length (`sequence_length`). Each sequence is essentially a window of past data points that the LSTM will use to make predictions.

3. **Target Data Adjustment**:
   - After reshaping the features, the target data (`y_train`) is adjusted accordingly by removing the initial sequences that do not have a corresponding target. This ensures that the features and targets align correctly in terms of time steps.

4. **Shape Verification**:
   - The shapes of the reshaped feature (`X_train_reshaped`) and target (`y_train_reshaped`) datasets are printed to verify that the data has been processed correctly. This step is crucial for ensuring that the data matches the expected input dimensions of the LSTM model.

5. **Feature Verification**:
   - The `check_feature_columns` function is called for each dataset (train, test, and extra) to verify that all necessary features, including lagged features and technical indicators like EMA, RSI, and MACD, are present. This ensures that no important features are missing, which could negatively impact model performance.

### Conclusion

By following this structured approach, I ensure that the training data is correctly reshaped for LSTM input and that all necessary features are present across the datasets. This preparation is essential for building a robust and accurate time series forecasting model. The code systematically verifies the data structure and content before proceeding to the modeling phase, minimizing the risk of errors during model training.


```python
# Set random seeds for reproducibility
set_random_seeds()


# Set sequence length (timestep)
sequence_length = 10

def reshape_for_lstm(X_data, sequence_length):
    """
    Reshape the data into sequences for LSTM input.

    Args:
    - X_data: Input features, shape (num_samples, num_features)
    - sequence_length: Number of time steps per sequence.

    Returns:
    - X_reshaped: Reshaped data, shape (num_sequences, sequence_length,
      num_features).
    """
    num_samples = X_data.shape[0]
    num_features = X_data.shape[1]

    # Calculate the number of sequences that can be created
    num_sequences = num_samples - sequence_length + 1

    # Initialize the reshaped array
    X_reshaped = np.zeros((num_sequences, sequence_length, num_features))

    # Populate the reshaped array with sequences
    for i in range(num_sequences):
        X_reshaped[i] = X_data[i:i + sequence_length]

    return X_reshaped

def check_feature_columns(data, ticker, n_lags=5):
    """
    Check if all expected features, including lagged features and
     technical indicators, are present.

    Args:
    - data: Dictionary containing feature data for each ticker.
    - ticker: String representing the stock ticker.
    - n_lags: Integer indicating the number of lagged features.

    Returns:
    - None. Prints missing features if any.
    """
    expected_features = [
        f'{ticker}_lag_{i}' for i in range(1, n_lags + 1)] + ['EMA_20', 'RSI', 'MACD']
    actual_features = list(data['X'].columns)

    missing_features = [feature for feature in expected_features\
                        if feature not in actual_features]
    if missing_features:
        print(f"Missing features for {ticker}: {missing_features}")
    else:
        print(f"All expected features present for {ticker}")

```

```python
# Set random seeds for reproducibility
set_random_seeds()

def prepare_data():
    """
    Prepare and preprocess the data for LSTM model.

    Returns:
    - X_train_reshaped: Reshaped input features for LSTM
    - y_train_reshaped: Corresponding target values
    """

    # Set sequence length (timestep)
    sequence_length = 10

    # Define directories
    out_dir_train = 'output/train_data'
    out_dir_test = 'output/test_data'
    out_dir_extra = 'output/extra_data'

    # Load training data from CSV
    train_data = pd.read_csv(os.path.join(out_dir_train, 'train_data.csv'),
                             index_col='Date')

    # Load test data from CSV
    test_data = pd.read_csv(os.path.join(out_dir_test, 'test_data.csv'),
                            index_col='Date')

    # Load extra test data from CSV
    extra_data = pd.read_csv(os.path.join(out_dir_extra, 'extra_data.csv'),
                             index_col='Date')

    # Preprocess datasets
    preprocessed_data = preprocess_data(train_data, test_data, extra_data)

    # Combine preprocessed train datasets
    X_train, y_train = combine_hist_train_datasets(preprocessed_data)

    # Reshape the training data
    X_train_reshaped = reshape_for_lstm(X_train, sequence_length)
    y_train_reshaped = y_train[sequence_length - 1:]

    return X_train_reshaped, y_train_reshaped

if __name__ == "__main__":
    X_train_reshaped, y_train_reshaped = prepare_data()

    print("Data preparation completed.")
    print(f"X_train_reshaped shape: {X_train_reshaped.shape}")
    print(f"y_train_reshaped shape: {y_train_reshaped.shape}")

    print("\nFirst few rows of X_train_reshaped:")
    print(X_train_reshaped[:3])

    print("\nFirst few values of y_train_reshaped:")
    print(y_train_reshaped[:10])

    print("\nData statistics:")
    print("X_train_reshaped mean:", np.mean(X_train_reshaped))
    print("X_train_reshaped std:", np.std(X_train_reshaped))
    print("y_train_reshaped mean:", np.mean(y_train_reshaped))
    print("y_train_reshaped std:", np.std(y_train_reshaped))
```

## Building and Compiling the LSTM Model

In this section, I focus on constructing and compiling a Long Short-Term Memory (LSTM) model, designed for time series forecasting. The model incorporates advanced features such as a bidirectional LSTM layer and an attention mechanism to enhance its performance in capturing temporal dependencies.

### Function Overview

The core function in this section is `build_lstm_model`, which constructs the LSTM model architecture. The `main` function orchestrates the model building process, including visualizing and summarizing the model architecture.

     - **Input Layer**: Defines the shape of the input data.

     - **Bidirectional LSTM Layer**: A key component that processes input sequences in both forward and backward directions, capturing long-term dependencies more effectively.

     - **Dropout Layer**: Applied to the LSTM layer to prevent overfitting by randomly dropping neurons during training.

     - **Attention Mechanism**: This layer allows the model to focus on the most relevant parts of the input sequence, improving its ability to capture important features.

     - **Concatenation**: Combines the output of the LSTM layer with the attention mechanism to enrich the feature representation.

     - **Flatten Layer**: Flattens the output to prepare it for the dense layers.

     - **Dense Layer with ReLU Activation**: Adds non-linearity to the model, allowing it to learn complex patterns.

     - **Dropout on Dense Layer**: Further regularizes the model by applying dropout to the dense layer.
     
     - **Output Layer**: A single neuron that produces the final prediction for regression tasks.

   - The `main` function returns the compiled LSTM model, which is ready for training and further use in the time series forecasting tasks.

### Conclusion

By following this structured approach, I successfully build a robust LSTM model tailored for time series forecasting. The model's architecture, which includes a bidirectional LSTM and attention mechanism, is designed to capture complex temporal patterns in the data, making it well-suited for predicting future trends. The model is fully compiled and ready for the next steps in the machine learning pipeline.


```python
# Set random seed reproducity
set_random_seeds


def build_lstm_model(input_shape):
    """
    Build and compile the LSTM model.

    Args:
    - input_shape: Shape of input data.

    Returns:
    - model: Compiled LSTM model.
    """
    # Define the input layer
    inputs = Input(shape=input_shape)

    # Add a bidirectional LSTM layer
    x = Bidirectional(LSTM(50,
                return_sequences=True))(inputs)

    # Apply dropout
    x = Dropout(0.2)(x)

    # Add attention mechanism
    attention = Attention()([x, x])

    # Concatenate LSTM and attention outputs
    x = Concatenate()([x, attention])

    # Flatten the output
    x = Flatten()(x)

    # Add dense layer with ReLU activation
    x = Dense(32, activation='relu')(x)

    # Apply dropout to dense layer
    x = Dropout(0.2)(x)

    # Define the output layer for regression
    outputs = Dense(1)(x)

    # Create and compile the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer=Adam(
        learning_rate=0.0005), loss='mse', metrics=['mae'])

    return model

def main():
    """
    Main function to build, visualize, and summarize the LSTM model.

    Returns:
    - model: Compiled STM model.
    """
    # Define Input shape
    input_shape = X_train_reshaped.shape[1:]

    # Build the LSTM model
    model = build_lstm_model(input_shape)

    # Plot and save the model architecture
    plot_model(model, to_file='lstm_model.png',
               show_shapes=True)
    display(Image(filename='lstm_model.png'))

    # Display a summary of the model
    model.summary()

    return model


if __name__ == "__main__":
    main()




```

In this section, I focus on evaluating the performance of the Long Short-Term Memory (LSTM) model on both test and extra test datasets.
```python
# Set random seeds for reproducibility
set_random_seeds()


def evaluate_model(model, X, y, scaler=None):
    """
    Evaluate model performance.

    Args:
    - model: Trained LSTM model.
    - X: Input features.
    - y: True target values.
    - scaler: Scaler to reverse scaling.

    Returns:
    - Dictionary with MAE, RMSE, R².
    """
    predictions = model.predict(X)

    if scaler is not None:
        y = scaler.inverse_transform(y.reshape(-1, 1)).flatten()
        predictions = scaler.inverse_transform(predictions).flatten()

    mae = mean_absolute_error(y, predictions)
    rmse = np.sqrt(mean_squared_error(y, predictions))
    r2 = r2_score(y, predictions)

    return {"MAE": mae, "RMSE": rmse, "R²": r2}

def plot_metrics(metrics, title):
    """
    Plot MAE, RMSE, R² values.

    Args:
    - metrics: Evaluation metrics dict.
    - title: Title for the plot.

    Returns:
    - None. Displays the plot.
    """
    fig, ax = plt.subplots(1, 3, figsize=(15, 5))
    ax[0].bar(metrics.keys(), [m["MAE"] for m in metrics.values()])
    ax[0].set_title('MAE')
    ax[1].bar(metrics.keys(), [m["RMSE"] for m in metrics.values()])
    ax[1].set_title('RMSE')
    ax[2].bar(metrics.keys(), [m["R²"] for m in metrics.values()])
    ax[2].set_title('R²')
    fig.suptitle(title)
    plt.savefig(f"{title}_metrics.png")
    plt.show()

def save_as_word_table(df, filename="evaluation_metrics.docx"):
    """
    Save the DataFrame as a Word table.

    Args:
    - df: DataFrame containing the metrics.
    - filename: Name of the Word file to save.
    """
    from docx import Document
    doc = Document()
    doc.add_heading('Evaluation Metrics', 0)

    # Add table to document
    table = doc.add_table(rows=(df.shape[0] + 1), cols=df.shape[1] + 1)

    # Adding headers
    hdr_cells = table.rows[0].cells
    hdr_cells[0].text = "Dataset/Ticker"
    for i, column_name in enumerate(df.columns):
        hdr_cells[i + 1].text = column_name

    # Adding data
    for i, idx in enumerate(df.index):
        row_cells = table.rows[i + 1].cells
        row_cells[0].text = str(idx)
        for j, value in enumerate(df.loc[idx]):
            row_cells[j + 1].text = str(value)

    # Save the document
    doc.save(filename)

def main():
    """
    Main function to evaluate and plot.

    Returns:
    - None. Prints and plots metrics.
    """
    model = load_model('lstm_model.keras')
    all_metrics = {}

    # Evaluate on test dataset
    print("=== Test Data ===")
    test_metrics = {}
    for ticker, data in preprocessed_data['test'].items():
        X_test_reshaped = reshape_for_lstm(data['X'].values, sequence_length)
        y_test_reshaped = data['y'].values[sequence_length - 1:]
        scaler = data.get('target_scaler')
        metrics = evaluate_model(model, X_test_reshaped, y_test_reshaped, scaler)
        test_metrics[ticker] = metrics
        print(f"{ticker}: {metrics}")
    all_metrics["Test Data"] = test_metrics

    # Evaluate on extra test dataset
    print("\n=== Extra Test Data ===")
    extra_metrics = {}
    for ticker, data in preprocessed_data['extra'].items():
        X_extra_reshaped = reshape_for_lstm(data['X'].values, sequence_length)
        y_extra_reshaped = data['y'].values[sequence_length - 1:]
        scaler = data.get('target_scaler')
        metrics = evaluate_model(model, X_extra_reshaped, y_extra_reshaped,
                                 scaler)
        extra_metrics[ticker] = metrics
        print(f"{ticker}: {metrics}")
    all_metrics["Extra Test Data"] = extra_metrics

    # Create DataFrame for results
    df = pd.DataFrame.from_dict(
        {(i, j): all_metrics[i][j]
         for i in all_metrics.keys()
         for j in all_metrics[i].keys()},
        orient='index')

    # Save and display table
    df.to_csv('evaluation_metrics.csv')
    save_as_word_table(df, "evaluation_metrics.docx")
    print(df)  # Display DataFrame directly

    # Plot and save metrics
    plot_metrics(test_metrics, "Test Data")
    plot_metrics(extra_metrics, "Extra Test Data")

if __name__ == "__main__":
    main()
```

```python
# Set random seeds for reproducibility
set_random_seeds()
```

```python
def train_xgb_model(X_train, X_val, y_train, y_val, params=None, num_boost=100,
                    seed_value=42):
    """
    Train the XGBoost model using the provided training and validation data.

    Args:
    - X_train: Training features.
    - X_val: Validation features.
    - y_train: Training targets.
    - y_val: Validation targets.
    - params: XGBoost parameters.
    - num_boost: Number of boosting rounds.
    - seed_value: Seed for reproducibility.

    Returns:
    - model: Trained XGBoost model.
    - evals_result: Dictionary with training and validation metrics.
    """
    # Ensure reproducibility
    set_random_seeds(seed_value)

    if params is None:
        params = {
            'objective': 'reg:squarederror',
            'eval_metric': ['mae', 'rmse'],
            'learning_rate': 0.1,
            'max_depth': 6,
            'seed': seed_value
        }

    dtrain = xgb.DMatrix(X_train, label=y_train)
    dval = xgb.DMatrix(X_val, label=y_val)

    evals_result = {}

    model = xgb.train(params, dtrain, num_boost, evals=[(dtrain, 'train'),
     (dval, 'val')], evals_result=evals_result, early_stopping_rounds=10,
                      verbose_eval=True)

    # Plot training MAE and validation MSE
    plt.figure(figsize=(12, 6))

    # Plot Training MAE (as in LSTM)
    plt.subplot(1, 2, 1)
    plt.plot(evals_result['train']['mae'], label='Training MAE')
    plt.title('Training MAE Over Rounds')
    plt.xlabel('Rounds')
    plt.ylabel('MAE')
    plt.legend()

    # Plot Validation MSE (interpreted from RMSE)
    plt.subplot(1, 2, 2)
    plt.plot(np.square(evals_result['val']['rmse']), label='Validation MSE')
    plt.title('Validation MSE Over Rounds')
    plt.xlabel('Rounds')
    plt.ylabel('MSE')
    plt.legend()

    plt.show()

    return model

def main_train():
    """
    Train XGBoost model and plot training progress.
    """
    X_train_split, X_val_split, y_train_split, y_val_split =\
     prepare_data_for_xgboost(X_train, y_train)

    model = train_xgb_model(X_train_split, X_val_split, y_train_split,
                            y_val_split, seed_value=42)

    # Save the trained model
    model.save_model('xgb_model.json')

if __name__ == "__main__":
    main_train()
```

## Training the Support Vector Machine (SVM) Model

In this section, I describe the process of training a Support Vector Machine (SVM) model, which is specifically configured to handle regression tasks. The training procedure involves preparing the data, fitting the model, and evaluating its performance using appropriate metrics. This approach ensures that the model is both accurate and generalizable.

### Workflow

1. **Ensuring Reproducibility**:

   - The core of this workflow is the `train_svm_model()` function, which is responsible for training the SVM model and evaluating its performance on both the training and validation datasets. The key components of this process include:

     - **Model Initialization**:
     
     I initialize the SVM model using the specified kernel type (`rbf` by default), regularization parameter (C), and epsilon for the epsilon-SVR model. These parameters control the complexity of the model and its ability to capture the underlying patterns in the data.

     - **Model Fitting**:
     
      The model is trained on the provided training data (`X_train` and `y_train`). During this step, the model learns to map the input features to the target values, optimizing its parameters to minimize the prediction error.

     - **Performance Evaluation**:
     
     After training, I evaluate the model's performance by making predictions on both the training and validation datasets. I calculate key metrics such as Mean Absolute Error (MAE) and Mean Squared Error (MSE) to assess how well the model performs. These metrics are displayed to provide immediate feedback on the model’s accuracy.

     - **Visualization**:
     
     To facilitate a clear understanding of the model's performance, I plot the MAE and MSE for both the training and validation datasets. These visualizations help identify any discrepancies between the training and validation performance, which could indicate overfitting or underfitting.

3. **Execution of the Main Training Function**:

   - The `main_train_svm()` function coordinates the overall process, ensuring that the model is trained and evaluated effectively:
     - **Data Preparation**: Initially, I combine the preprocessed datasets into a single training set. This step ensures that the model is trained on a comprehensive dataset that captures a wide range of patterns and behaviors.
     
     - **Data Splitting**:
     
      The combined dataset is then split into training and validation sets using an 80/20 split. This ensures that the model's performance is evaluated on data it has not seen during training, providing a more accurate measure of its generalization ability.

     - **Model Training and Evaluation**:
     
     Finally, I train the SVM model using the training data and evaluate its performance on the validation data. The function also generates plots to visualize the training progress and the accuracy of the model.

### Conclusion

By following this structured approach, I ensure that the SVM model is trained efficiently and evaluated comprehensively. The use of performance metrics and visualizations allows for a detailed assessment of the model's accuracy and generalization capabilities, which are critical for making informed decisions about model deployment in real-world applications.


```python
# Set random seeds for reproducibility
set_random_seeds()

def train_svm_model(X_train, y_train, X_val, y_val, kernel='rbf', C=1.0,
                    epsilon=0.1):
    """
    Train an SVM model using the provided training and validation data,
    display the training metrics, and plot the metrics.

    Args:
    - X_train: Training features.
    - y_train: Training targets.
    - X_val: Validation features.
    - y_val: Validation targets.
    - kernel: Kernel type to be used in the algorithm.
    - C: Regularization parameter.
    - epsilon: Epsilon in the epsilon-SVR model.

    Returns:
    - model: Trained SVM model.
    """
    model = SVR(kernel=kernel, C=C, epsilon=epsilon)
    model.fit(X_train, y_train)

    # Predictions on training and validation data
    train_predictions = model.predict(X_train)
    val_predictions = model.predict(X_val)

    # Calculate metrics
    train_mae = mean_absolute_error(y_train, train_predictions)
    train_mse = mean_squared_error(y_train, train_predictions)
    val_mae = mean_absolute_error(y_val, val_predictions)
    val_mse = mean_squared_error(y_val, val_predictions)

    # Display the training and validation metrics
    print("Training Metrics:")
    print(f"Training MAE: {train_mae:.4f}")
    print(f"Training MSE: {train_mse:.4f}")

    print("\nValidation Metrics:")
    print(f"Validation MAE: {val_mae:.4f}")
    print(f"Validation MSE: {val_mse:.4f}")

    # Plotting the training and validation metrics
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.bar(['Training MAE', 'Validation MAE'], [train_mae, val_mae])
    plt.title('MAE on Training and Validation Data')
    plt.ylabel('MAE')

    plt.subplot(1, 2, 2)
    plt.bar(['Training MSE', 'Validation MSE'], [train_mse, val_mse])
    plt.title('MSE on Training and Validation Data')
    plt.ylabel('MSE')

    # Explicitly show the plots
    plt.show()

    return model

def main_train_svm():
    """
    Main function to combine datasets, split the data, train the SVM model,
    and plot the training metrics.

    This function performs the following tasks:
    1. Combines the preprocessed datasets into a single training set.
    2. Splits the data into training and validation sets.
    3. Trains the SVM model using the training data.
    4. Plots the training and validation metrics.

    Returns:
    - model: Trained SVM model.
    """
    # Combine the datasets
    X_train, y_train = combine_datasets(preprocessed_data)

    # Split the data into training and validation sets
    X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
        X_train, y_train, test_size=0.2, random_state=42)

    # Train the SVM model and plot training metrics
    model = train_svm_model(X_train_split, y_train_split, X_val_split, y_val_split)

    return model

if __name__ == "__main__":
    model = main_train_svm()
```
