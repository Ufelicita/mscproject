## Setting Random Seeds for Reproducibility

In this section, I focus on setting random seeds across various libraries to ensure the reproducibility of the results. Consistent initialization of random seeds is crucial in machine learning and data science projects to ensure that experiments yield consistent and comparable results across different runs.

### Function Overview

The primary goal of this section is to define and implement a function that configures random seeds for different random number generators (RNGs) used in the project. By setting these seeds, I ensure that the random processes involved in data splitting, model initialization, and training are consistent, which is essential for debugging and replicating results.

### Workflow

1. **Defining the Function to Set Random Seeds**:
   - The `set_random_seeds` function is designed to initialize random seeds for multiple libraries, including NumPy, Python's `random` module, TensorFlow, and XGBoost. This function takes an optional `seed_value` parameter, allowing flexibility in the choice of the seed value, with a default set to 42.

   - **Inputs**:
     - `seed_value`: The seed value to be used for RNGs across libraries. The default value is set to 42, a commonly used seed value in data science for consistency.

   - **Outputs**:
     - None. The function sets seeds within the global environment, ensuring reproducibility.

2. **Setting Seeds Across Libraries**:
   - **NumPy**: The function sets the seed for NumPy's random number generator using `np.random.seed(seed_value)`. NumPy is widely used for numerical operations, and setting its RNG seed is important for consistency in operations that involve randomness, such as data shuffling.
   
   - **Python `random` Module**: The seed for Python's built-in `random` module is set using `random.seed(seed_value)`. This ensures that any operations relying on Python's RNG, such as random sampling or shuffling, are reproducible.

   - **TensorFlow**: TensorFlow's RNG seed is set using `tf.random.set_seed(seed_value)`. TensorFlow is a major library for deep learning, and setting its RNG seed ensures consistent initialization of model weights and operations that involve randomness.

   - **XGBoost**:
     - The verbosity of XGBoost is configured to be silent using `xgb.set_config(verbosity=0)`, which reduces unnecessary output during training.
     - The seed for XGBoost's RNG is set in its parameters using `xgparam = {'seed': seed_value}`. This ensures that any randomness within XGBoost models, such as during tree construction, is consistent.

3. **Initializing Random Seed for Consistency**:
   - After defining the function, I call `set_random_seeds()` to initialize the random seed with the default value of 42. This step is crucial for ensuring that all subsequent operations in the script are performed with a consistent seed, thus enabling reproducibility.

### Conclusion

By setting random seeds across multiple libraries, I ensure that the results of the project are reproducible. This is a critical step in any data science or machine learning workflow, as it allows for consistent experimentation and reliable comparison of results. The `set_random_seeds` function provides a centralized and easy-to-use method for achieving this consistency.


```python
# Function to set random seeds for reproducibility
def set_random_seeds(seed_value=42):
    """
    Configure random seed for reproducibility.
    Args:
    - seed_value: Seed for RNGs (default: 42).
    Returns:
    - None
    """
    # Set seed for numpy RNG
    np.random.seed(seed_value)

    # Set seed for Python's random RNG
    random.seed(seed_value)

    # Set seed for TensorFlow RNG
    tf.random.set_seed(seed_value)

    # Configure XGBoost verbosity to 0 (silent)
    xgb.set_config(verbosity=0)

    # Set seed for XGBoost parameters
    xgparam = {'seed': seed_value}

# Initialise random seed for consistency
set_random_seeds()
```

## Downloading and Processing Stock Data

In this section, I focus on downloading historical stock data for specified tickers, processing the data to retain only the adjusted close prices, and combining the processed data into unified datasets for training, testing, and additional analysis. This step is critical for preparing the data in a format suitable for time-series modeling and further analysis.

### Function Overview

The primary goal of this section is to download and process stock data for different sets of tickers over specified date ranges. The data is processed to retain only the adjusted close prices, renamed according to the respective ticker symbols, and combined into a single DataFrame for each dataset. Additionally, information about the number of data points before and after processing is recorded and displayed.

### Workflow

1. **Fetching Stock Data**:
   - The `fetch_data` function is responsible for downloading the raw stock data for a given ticker over a specified date range. The data is saved to a CSV file if it doesn't already exist in the output directory.

   - **Inputs**:
     - `ticker`: The stock ticker symbol to download data for.
     - `start`: The start date for the data download (format: 'YYYY-MM-DD').
     - `end`: The end date for the data download (format: 'YYYY-MM-DD').
     - `output_dir`: The directory where the downloaded data will be saved.

   - **Outputs**:
     - `data`: The downloaded data as a pandas DataFrame.
     - `row_count`: The number of rows in the downloaded data.

   - The function ensures that the output directory exists, checks if the data file already exists, and either downloads the data or loads it from the existing file. It returns the data along with the number of rows in the dataset.

2. **Processing Stock Data**:
   - The `process_data` function processes the raw stock data by keeping only the 'Date' and 'Adj Close' columns and renaming the 'Adj Close' column to the respective ticker symbol. This step standardizes the data, making it easier to combine and analyze.

   - **Inputs**:
     - `df`: The raw stock data as a pandas DataFrame.
     - `ticker`: The stock ticker symbol, used for renaming the 'Adj Close' column.

   - **Outputs**:
     - `df`: The processed data with the 'Adj Close' column renamed to the ticker symbol.
     - `row_count`: The number of rows in the processed data.

   - The function converts the 'Date' column to datetime format and keeps only the relevant columns. The processed data is returned along with the number of rows, which is useful for tracking how much data remains after processing.

3. **Downloading and Processing Data for Multiple Tickers**:
   - The `download_and_process_data` function orchestrates the download and processing of data for multiple tickers. It combines the processed data from all tickers into a single DataFrame, aligned by date, and records the number of data points before and after processing for each ticker.

   - **Inputs**:
     - `tickers`: A list of stock ticker symbols to process.
     - `start`: The start date for the


```python
def fetch_data(ticker, start, end, output_dir):
    """
    Fetch and save stock data for a ticker.
    Args:
    - ticker (str): Stock ticker symbol.
    - start (str): Start date (YYYY-MM-DD).
    - end (str): End date (YYYY-MM-DD).
    - output_dir (str): Directory to save data.
    Returns:
    - pd.DataFrame: Loaded data for ticker.
    - int: Number of rows in the downloaded data.
    """
    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Define file path for the data
    file_path = os.path.join(output_dir, f"{ticker}.csv")

    # Download data
    if not os.path.exists(file_path):
        data = yf.download(ticker, start=start, end=end)
        data.reset_index(inplace=True)  # Ensure 'Date' is a column
        data.to_csv(file_path, index=False)  # Save without index
    else:
        data = pd.read_csv(file_path)

    # Return the data as DataFrame and its row count
    return data, data.shape[0]

# Function to process data
def process_data(df, ticker):
    """
    Process data: keep 'Adj Close', rename.
    Args:
    - df (pd.DataFrame): Stock data.
    - ticker (str): Stock ticker symbol.
    Returns:
    - pd.DataFrame: Processed data with 'Adj Close'.
    - int: Number of rows after processing.
    """
    # Convert 'Date' to datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Keep only 'Date' and 'Adj Close' columns
    df = df[['Date', 'Adj Close']]

    # Rename 'Adj Close' to ticker symbol
    df = df.rename(columns={'Adj Close': ticker})

    # Return the processed DataFrame and its row count
    return df, df.shape[0]

# Function to download and process data
def download_and_process_data(tickers, start, end, out_dir):
    """
    Download and process stock data.
    Args:
    - tickers (list): List of stock tickers.
    - start (str): Start date (YYYY-MM-DD).
    - end (str): End date (YYYY-MM-DD).
    - out_dir (str): Directory to save data.
    Returns:
    - pd.DataFrame: Combined data for tickers.
    - dict: Dictionary of ticker with data points before and after processing.
    """
    # Initialize empty DataFrame
    combined_df = None

    # Initialize a dictionary to store data points information
    data_points_info = {}

    # Loop through each ticker
    for ticker in tickers:
        # Fetch raw data
        raw_df, raw_count = fetch_data(ticker, start, end, out_dir)

        # Process data for current ticker
        processed_df, processed_count = process_data(raw_df,
                                                     ticker.split('.')[0])

        # Record the data points information
        data_points_info[ticker] = {
            'before_processing': raw_count,
            'after_processing': processed_count
        }

        # Merge processed data with combined_df
        combined_df = (processed_df if combined_df is None
                       else pd.merge(combined_df, processed_df,
                                     on='Date', how='inner'))

    # Return the combined DataFrame and the data points info
    return combined_df, data_points_info

# Define tickers for each dataset
tickers_train = ['LLOY.L', 'SHEL.L', 'AZN.L', 'RIO.L', 'TSCO.L',
                 'VOD.L', 'BA.L', 'LAND.L', 'REL.L', 'SGE.L']
tickers_test = ['BARC.L', 'DGE.L', 'BP.L', 'GSK.L', 'GLEN.L', 'OCDO.L',
                'PRU.L', 'RR.L', 'BLND.L', 'ULVR.L']
tickers_extra =['LGEN.L', 'IMB.L', 'SSE.L', 'SN.L', 'AAL.L', 'SBRY.L',
                'WOOD.L', 'WEIR.L', 'SGRO.L', 'EXPN.L']

# Define date ranges for datasets
date_train = ('2014-01-01', '2023-12-31')
date_test = ('2024-01-01', '2024-03-31')
date_extra = ('2024-01-01', '2024-06-30')

# Directories for saving data
dirs = ('train_data', 'test_data', 'extra_data')

# Download and process training data
combined_train_data, train_data_points = \
download_and_process_data(tickers_train, *date_train, dirs[0])

# Download and process test data
combined_test_data, test_data_points =\
 download_and_process_data(tickers_test, *date_test, dirs[1])

# Download and process extra data
combined_extra_data, extra_data_points = \
download_and_process_data(tickers_extra, *date_extra, dirs[2])

# Print first few rows of each dataset
print(combined_train_data.head())
print(combined_test_data.head())
print(combined_extra_data.head())

# Print the data points information
print("Training Data Points:", train_data_points)
print("Test Data Points:", test_data_points)
print("Extra Data Points:", extra_data_points)
```

## Data Preprocessing for Stock Price Prediction

In this section, I focus on preprocessing stock price data to create lagged features and compute technical indicators. ### Function Overview

   - **Inputs**:

     - `n_lags`: The number of lagged features to create for each stock price series. Default is 5.
     - `output_dir`: The directory where the processed data and scalers will be saved. Default is 'processed_data'.

   - **Outputs**:
     - `preprocessed_data`: A dictionary containing the processed features and targets for each ticker, along with the corresponding feature and target scalers.

   - The function first ensures that the specified output directory exists. If it doesn't, the directory is created to store the processed data.

4. **Initializing Data Structures**:
   - A dictionary `preprocessed_data` is initialized to store the processed data for training, testing, and extra datasets. This structure keeps the data organized by dataset type and ticker symbol.

5. **Processing Each Dataset**:
   - The function iterates over the training, test, and extra datasets. For each dataset, it processes the data for each ticker symbol individually.

   - **Indexing by Date**:
     - If the data is not already indexed by date, it is converted to a datetime format and set as the index.

   - **Computing Technical Indicators**:
     - Technical indicators such as EMA, RSI, and MACD are computed for each stock. These indicators are commonly used in financial analysis and are added as features to the dataset.

   - After generating lagged features, rows with missing values (due to lagging) are dropped. If no data remains after this step, a message is printed, and the ticker is skipped.

   - The features and target variables are scaled using `RobustScaler` and `StandardScaler`, respectively. Scaling ensures that the data is normalized, which is essential for many machine learning algorithms.

8. **Storing and Saving Processed Data**:
   - The scaled features and targets, along with their respective scalers, are stored in the `preprocessed_data` dictionary. Additionally, the processed data and scalers are saved as CSV and pickle files, respectively, for future use.

9. **Final Data Shape Verification**:
   - The shapes of the processed datasets are printed to verify that the preprocessing was completed successfully. This step helps ensure that the data is in the expected format before proceeding to model training.

This standardized preprocessing workflow not only enhances the quality of the input data but also ensures that the models can effectively learn patterns from historical stock prices. The resulting datasets are saved for future use, streamlining the process of model development and evaluation.


```python
# Set random seeds for reproducibility
set_random_seeds()

def preprocess_data(train_data, test_data, extra_data, n_lags=5,
                    output_dir='processed_data'):
    """
    Preprocess stock data, create lagged features, and save processed data.
    """
    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Initialize dictionary for data
    preprocessed_data = {'train': {}, 'test': {}, 'extra': {}}

    # Datasets to iterate through
    datasets = {'train': train_data, 'test': test_data,  'extra': extra_data}

    for name, data in datasets.items():
        print(f"Processing {name} data")

        if data.index.name != 'Date':
            print(f"Indexing {name} by 'Date'")
            data['Date'] = pd.to_datetime(data['Date'])
            data.set_index('Date', inplace=True)

        for ticker in data.columns:
            print(f"Processing {ticker}")
            stock_data = data[[ticker]].copy()

            # Compute technical indicators
            stock_data['EMA_20'] = ta.trend.ema_indicator(
                stock_data[ticker], window=20)
            stock_data['RSI'] = ta.momentum.rsi(
                stock_data[ticker], window=14)
            stock_data['MACD'] = ta.trend.macd(
                stock_data[ticker])

            # Create lagged features
            for lag in range(1, n_lags + 1):
                stock_data[f'{ticker}_lag_{lag}'] = \
                    stock_data[ticker].shift(lag)

            stock_data.dropna(inplace=True)

            if stock_data.empty:
                print(f"No data left for {ticker}")
                continue

            # Separate features and target
            X = stock_data.drop(columns=[ticker])
            y = stock_data[ticker]

            # Scale features and target
            feature_scaler = RobustScaler()
            target_scaler = StandardScaler()
            X_scaled = feature_scaler.fit_transform(X)
            y_scaled = target_scaler.fit_transform(
                y.values.reshape(-1, 1)).flatten()

            X_scaled = pd.DataFrame(X_scaled, columns=X.columns,
                                    index=X.index)
            y_scaled = pd.Series(y_scaled, index=y.index)

            # Store scaled data
            preprocessed_data[name][ticker] = {
                'X': X_scaled,
                'y': y_scaled,
                'feature_scaler': feature_scaler,
                'target_scaler': target_scaler
            }

            # Save processed data
            X_scaled.to_csv(os.path.join(output_dir,
                              f'{name}_{ticker}_X.csv'))
            y_scaled.to_csv(os.path.join(output_dir,
                              f'{name}_{ticker}_y.csv'))
            joblib.dump(feature_scaler, os.path.join(output_dir,
                        f'{name}_{ticker}_feature_scaler.pkl'))
            joblib.dump(target_scaler, os.path.join(output_dir,
                        f'{name}_{ticker}_target_scaler.pkl'))

    # Print data shapes
    print("\nFinal shapes of preprocessed data:")
    for name, dataset in preprocessed_data.items():
        print(f"\n{name.capitalize()} data:")
        for ticker, data in dataset.items():
            print(f"  {ticker}: X shape {data['X'].shape}, "
                  f"y shape {data['y'].shape}")

    return preprocessed_data


# Load datasets from CSV
train_data = pd.read_csv('output/train_data/train_data.csv')
test_data = pd.read_csv('output/test_data/test_data.csv')
extra_data = pd.read_csv('output/extra_data/extra_data.csv')

# Preprocess the datasets
preprocessed_data = preprocess_data(train_data, test_data, extra_data)

```

```python
# Set random seeds for reproducibility
set_random_seeds()

def combine_hist_train_datasets(preprocessed_data):
    """
    Combine historical training datasets.

    Args:
    - preprocessed_data: Dict with preprocessed data.

    Returns:
    - combined_X: Combined input features.
    - combined_y: Combined target values.

    This function combines the features (X) and target
    values (y) from all tickers in the training data.
    """
    combined_X = []  # List to hold combined features
    combined_y = []  # List to hold combined targets

    # Loop through each ticker's training data
    for ticker, data in preprocessed_data['train'].items():
        combined_X.append(data['X'].values)  # Append feature data
        combined_y.append(data['y'].values)  # Append target data

    # Vertically stack features from all tickers
    combined_X = np.vstack(combined_X)
    # Concatenate target values into one array
    combined_y = np.concatenate(combined_y)

    return combined_X, combined_y


# Call the function with the example data
combined_X, combined_y = combine_hist_train_datasets(preprocessed_data)

# Output the shapes of the combined data
print(f"Combined X shape: {combined_X.shape}")
print(f"Combined y shape: {combined_y.shape}")
```

## Preparing Data for LSTM Model

In this section, I focus on preparing the input data for an LSTM (Long Short-Term Memory) model. This involves setting random seeds for reproducibility, reshaping the data into a format suitable for LSTM, and saving the reshaped data for future use. The steps taken here are critical for ensuring that the model can efficiently learn from the time-series data.

### Function Overview

The primary purpose of this section is to reshape the input data so that it can be used effectively by the LSTM model. LSTMs require input data in a 3D format, where each sample contains a sequence of timesteps. By restructuring the data accordingly, I ensure that the LSTM model can capture temporal dependencies within the dataset.

### Workflow

1. **Setting Random Seeds for Reproducibility**:
   - To maintain consistency across different runs, I start by setting random seeds. This ensures that the results are reproducible and not affected by random variations in the data processing steps or model training.

2. **Reshaping Data for LSTM**:
   - The `reshape_for_lstm` function is defined to reshape the input feature data into a 3D array suitable for LSTM models. This function takes the 2D input data and reorganizes it into sequences of a specified length (number of timesteps).

   - **Inputs**:
     - `X`: The input feature data, originally a 2D array.
     - `sequence_length`: The number of timesteps to include in each sequence.

   - **Outputs**:
     - `X_reshaped`: A 3D array where each sample now contains a sequence of timesteps. This reshaped data is suitable for feeding into an LSTM model.

3. **Loading the Combined Training Data**:
   - The combined training data is loaded from `.npy` files, which contain the input features (`X_train`) and target values (`y_train`).

4. **Setting Sequence Length (Timesteps)**:
   - A sequence length of 120 timesteps is set, which determines how many previous data points the LSTM will consider when making predictions.

5. **Reshaping the Training Data**:
   - The training data is reshaped using the `reshape_for_lstm` function. This step is crucial as it prepares the data in the required 3D format (samples, timesteps, features) for the LSTM model.
   - The target values (`y_train_reshaped`) are adjusted to match the reshaped input data by excluding the first `sequence_length - 1` data points.

6. **Saving the Reshaped Training Data**:
   - The reshaped input features (`X_train_reshaped`) and target values (`y_train_reshaped`) are saved to `.npy` files. This ensures that the preprocessed data is stored and can be easily loaded for model training later.

7. **Verifying the Structure of the Reshaped Data**:
   - Finally, the shapes of the reshaped data arrays are printed to verify that the reshaping process has been completed correctly. This step is important for confirming that the data is in the expected format before proceeding to model training.

### Conclusion

By setting random seeds, reshaping the data, and saving the preprocessed datasets, I ensure that the data is ready for efficient training of the LSTM model. This preparation step is critical for capturing the temporal dependencies in the data, which are essential for time-series forecasting tasks.


```python
# Set random seeds for reproducibility
set_random_seeds()


def reshape_for_lstm(X, sequence_length):
    """
    Reshape the input data for LSTM.

    Args:
    - X: Input feature data (2D array).
    - sequence_length: Number of timesteps to use.

    Returns:
    - X_reshaped: Reshaped input data for LSTM.

    The function creates sequences of data using a
    rolling window approach to match LSTM input shape.
    """
    X_reshaped = []  # List to hold reshaped sequences

    # Create sequences using a rolling window
    for i in range(len(X) - sequence_length + 1):
        X_reshaped.append(X[i:i + sequence_length])

    return np.array(X_reshaped)  # Convert to numpy array
```

## Building and Compiling the LSTM Model

In this section, I focus on constructing and compiling a Long Short-Term Memory (LSTM) model, designed for time series forecasting. The model incorporates advanced features such as a bidirectional LSTM layer and an attention mechanism to enhance its performance in capturing temporal dependencies.

### Function Overview

The core function in this section is `build_lstm_model`, which constructs the LSTM model architecture. The `main` function orchestrates the model building process, including visualizing and summarizing the model architecture.

### Workflow

1. **Setting Random Seeds**:
   - To ensure reproducibility of results, I start by setting random seeds using the `set_random_seeds()` function. This step is critical for consistent model training and evaluation across different runs.

     - **Input Layer**: Defines the shape of the input data.

     - **Bidirectional LSTM Layer**: A key component that processes input sequences in both forward and backward directions, capturing long-term dependencies more effectively.

     - **Dropout Layer**: Applied to the LSTM layer to prevent overfitting by randomly dropping neurons during training.

     - **Attention Mechanism**: This layer allows the model to focus on the most relevant parts of the input sequence, improving its ability to capture important features.

     - **Concatenation**: Combines the output of the LSTM layer with the attention mechanism to enrich the feature representation.

     - **Flatten Layer**: Flattens the output to prepare it for the dense layers.

     - **Dense Layer with ReLU Activation**: Adds non-linearity to the model, allowing it to learn complex patterns.

     - **Dropout on Dense Layer**: Further regularizes the model by applying dropout to the dense layer.
     
     - **Output Layer**: A single neuron that produces the final prediction for regression tasks.

3. **Model Compilation**:

   - The model is compiled with the Adam optimizer and Mean Squared Error (MSE) loss function, making it suitable for regression tasks. The Mean Absolute Error (MAE) is also tracked as a metric during training.

4. **Model Visualization and Summary**:
   - The model architecture is visualized using the `plot_model` function, which saves the architecture as an image (`lstm_model.png`). This visualization provides insight into the structure and complexity of the model.

5. **Returning the Model**:
   - The `main` function returns the compiled LSTM model, which is ready for training and further use in the time series forecasting tasks.

### Conclusion

By following this structured approach, I successfully build a robust LSTM model tailored for time series forecasting. The model's architecture, which includes a bidirectional LSTM and attention mechanism, is designed to capture complex temporal patterns in the data, making it well-suited for predicting future trends. The model is fully compiled and ready for the next steps in the machine learning pipeline.

   - To ensure reproducibility, random seeds are set using the `set_random_seeds()` function. This guarantees that the evaluation results are consistent across different runs.

2. **Model Evaluation**:
   - The `evaluate_model()` function is central to this process. It predicts the target values for the given input features and calculates the following metrics:

     - **MAE (Mean Absolute Error)**: Measures the average magnitude of errors in predictions, without considering their direction.

     - **RMSE (Root Mean Squared Error)**: Provides a measure of the differences between predicted and observed values, giving higher weight to large errors.

   - The `plot_metrics()` function visualizes the MAE, RMSE, and R² values using bar plots. These plots provide a quick and intuitive understanding of the model's performance across different datasets.

4. **Saving Results**:
   - The `save_as_word_table()` function is designed to save the evaluation metrics as a table in a Word document. This is particularly useful for documentation and reporting purposes.
   - Additionally, the metrics are saved as a CSV file for further analysis.

5. **Main Function**:
   - The `main()` function ties together all the components. It loads the trained LSTM model, evaluates it on both test and extra test datasets, saves the results, and generates visualizations.

```python
# Set random seeds for reproducibility
set_random_seeds()

# Custom RMSE metric
def rmse(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true)))

def build_lstm_model(hp):
    """
    Build and compile an LSTM model with hyperparameter tuning support.
    """
    inputs = Input(shape=(None, 8))  # Assuming 8 features

    # Tune the number of LSTM units
    lstm_units = hp.Int('lstm_units', min_value=32, max_value=128, step=32)

    # BiLSTM with increased regularization
    x = Bidirectional(LSTM(lstm_units, return_sequences=True,
                           kernel_regularizer=regularizers.l2(
                               hp.Float('regularization_strength',
                                        min_value=1e-4, # regularization higher
                                        max_value=1e-2,
                                        sampling='LOG'))))(inputs)

    # Dropout to prevent overfitting, with a possibility to increase the rate
    dropout_rate = hp.Float('dropout_rate', min_value=0.3,
                            max_value=0.7, step=0.1)  # Increased minimum dropout
    x = Dropout(dropout_rate)(x)

    # Attention mechanism for key features
    attention = Attention()([x, x])

    # Concatenate LSTM and attention outputs
    x = Concatenate()([x, attention])

    # Global Average Pooling layer
    x = GlobalAveragePooling1D()(x)

    # Dense layer with ReLU and regularization
    dense_units = hp.Int('dense_units', min_value=32,
                         max_value=128, step=32)  # Wider range for dense units
    x = Dense(dense_units, activation='relu',
              kernel_regularizer=regularizers.l2(
                  hp.Float('dense_regularization_strength',
                           min_value=1e-4,  # increase minimum regularization
                           max_value=1e-2,
                           sampling='LOG')))(x)

    # Dropout in dense layer
    x = Dropout(dropout_rate)(x)

    # Output layer for regression
    outputs = Dense(1)(x)

    # Compile model with Adam, RMSE, MAE
    learning_rate = hp.Float('learning_rate', min_value=1e-4, max_value=1e-2,
                             sampling='LOG')
    model = Model(inputs, outputs)
    model.compile(optimizer=Adam(learning_rate=learning_rate),
                  loss='mse', metrics=['mae', rmse])

    return model

def hyperparameter_tuning_and_training(X_train, y_train, X_val, y_val,
                                       tuner_directory='lstm_tuning',
                                       project_name='bayesian_optimization_lstm'):
    """
    Perform hyperparameter tuning, train the model with the best hyperparameters,
    save the trained model's weights, and return the trained model.
    """
    # Adjusted early stopping to potentially stop training earlier
    early_stopping = EarlyStopping(monitor='val_loss', patience=2,
                                   restore_best_weights=True)

    # Initialize the Bayesian Optimization tuner
    tuner = BayesianOptimization(
        build_lstm_model,
        objective='val_loss',  # Optimize based on validation loss (MSE)
        max_trials=20,
        executions_per_trial=1,
        directory=tuner_directory,
        project_name=project_name
    )

    # Start the hyperparameter search
    tuner.search(X_train, y_train,
                 epochs=50,
                 batch_size=32,
                 validation_data=(X_val, y_val),
                 callbacks=[early_stopping])

    # Get the best hyperparameters
    best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

    print(f"""
    The hyperparameter search is complete.
    Optimal LSTM units: {best_hps.get('lstm_units')}
    Optimal Dense units: {best_hps.get('dense_units')}
    Optimal Dropout rate: {best_hps.get('dropout_rate')}
    Optimal Regularization strength: {best_hps.get('regularization_strength')}
    Optimal Learning rate: {best_hps.get('learning_rate')}
    """)

    # Build the final model with the best hyperparameters
    ht_model = tuner.hypermodel.build(best_hps)

    # Train the model with the best hyperparameters
    history = ht_model.fit(X_train, y_train, epochs=50, batch_size=32,
                           validation_data=(X_val, y_val),
                           callbacks=[early_stopping])

    # Save the trained model
    ht_model.save('trained_ht_lstm_model.keras')
    print("Trained model saved to 'trained_ht_lstm_model.keras'")

    # Save the model weights
    weights_file = 'trained_ht_lstm_model.weights.h5'
    ht_model.save_weights(weights_file)
    print(f"Model weights saved to {weights_file}")

    # Verify that the weights have been saved
    if os.path.exists(weights_file):
        print("Weights file successfully created.")
    else:
        print("Error: Weights file not found.")

    # Save the model architecture
    model_json = ht_model.to_json()
    with open("trained_ht_lstm_model.json", "w") as json_file:
        json_file.write(model_json)
    print("Model architecture saved to trained_ht_lstm_model.json")

    # Plot training and validation loss
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    # Plot training and validation MAE
    plt.subplot(1, 2, 2)
    plt.plot(history.history['mae'], label='Train MAE')
    plt.plot(history.history['val_mae'], label='Val MAE')
    plt.title('Model MAE')
    plt.xlabel('Epoch')
    plt.ylabel('MAE')
    plt.legend()

    plt.show()

    # Display model summary
    ht_model.summary()

    # Visualize and save the model architecture
    plot_model(ht_model, to_file='ht_lstm_model_architecture.png',
               show_shapes=True, show_layer_names=True)
    display(Image(filename='ht_lstm_model_architecture.png'))

    # Return the model, the path to the weights file, and best hyperparameters
    return ht_model, weights_file, best_hps

# Call the hyperparameter tuning function
ht_model, weights_file, best_hps = hyperparameter_tuning_and_training(
    X_train, y_train, X_val, y_val
)
```

## Training the XGBoost Model

In this section, I outline the procedure for training an XGBoost model, which I have carefully tailored to effectively manage the prepared dataset. The training process involves the thoughtful initialization of model parameters, iterative learning, and comprehensive performance evaluation to ensure that the model generalizes well to new, unseen data.

### Workflow

1. **Ensuring Reproducibility**:
   - To guarantee the reproducibility of my experimental results, I begin by setting random seeds using the `set_random_seeds()` function. This step is crucial to ensure consistent outcomes across different runs of the model, thereby enhancing the reliability and validity of my findings.

2. **Model Training Process**:
   - The core of this process is the `train_xgb_model()` function, which I use to orchestrate the training of the XGBoost model. This function handles several key aspects of the training process:
     - **Parameter Initialization**: In cases where specific parameters are not provided, I initialize a default set of hyperparameters. These include the learning objective (`reg:squarederror`), evaluation metrics (Mean Absolute Error (MAE) and Root Mean Squared Error (RMSE)), learning rate, and maximum tree depth. These parameters are critical in guiding the model's learning process and ensuring that it can capture the underlying patterns in the data.
     - **Training Procedure**: I then proceed with the training of the model using the training dataset. During this process, I monitor the model's performance on a validation set. I employ early stopping to prevent overfitting by halting the training if there is no improvement in validation metrics after a set number of iterations. This strategy not only safeguards against overfitting but also conserves computational resources.
     - **Performance Evaluation**: Throughout the training process, I evaluate the model's performance by tracking the Mean Absolute Error (MAE) on the training data and the Mean Squared Error (MSE) on the validation data, derived from the RMSE. I plot these metrics across the training rounds, which helps me visually assess the model's learning curve and its ability to generalize.

3. **Executing the Main Training Function**:
   - Finally, in the `main_train()` function, I coordinate the overall training workflow:
     - **Data Preparation**: I start by preparing the training and validation datasets by splitting the input data. This ensures that the model is trained on a robust dataset while also validating its performance on a separate validation set.

The use of early stopping, rigorous evaluation, and proper parameter initialization helps me achieve optimal results, making the model well-suited for predictive modeling tasks in my research.


```python
# Set random seeds for reproducibility
set_random_seeds()

def train_xgb_model(X_train, X_val, y_train, y_val, params=None, num_boost=100,
                    seed_value=42):
    """
    Train the XGBoost model using the provided training and validation data.

    Args:
    - X_train: Training features.
    - X_val: Validation features.
    - y_train: Training targets.
    - y_val: Validation targets.
    - params: XGBoost parameters.
    - num_boost: Number of boosting rounds.
    - seed_value: Seed for reproducibility.

    Returns:
    - model: Trained XGBoost model.
    - evals_result: Dictionary with training and validation metrics.
    """
    # Ensure reproducibility
    set_random_seeds(seed_value)

    if params is None:
        params = {
            'objective': 'reg:squarederror',
            'eval_metric': ['mae', 'rmse'],
            'learning_rate': 0.1,
            'max_depth': 6,
            'seed': seed_value
        }

    dtrain = xgb.DMatrix(X_train, label=y_train)
    dval = xgb.DMatrix(X_val, label=y_val)

    evals_result = {}

    model = xgb.train(params, dtrain, num_boost, evals=[(dtrain, 'train'),
     (dval, 'val')], evals_result=evals_result, early_stopping_rounds=10,
                      verbose_eval=True)

    # Plot training MAE and validation MSE
    plt.figure(figsize=(12, 6))

    # Plot Training MAE (as in LSTM)
    plt.subplot(1, 2, 1)
    plt.plot(evals_result['train']['mae'], label='Training MAE')
    plt.title('Training MAE Over Rounds')
    plt.xlabel('Rounds')
    plt.ylabel('MAE')
    plt.legend()

    # Plot Validation MSE (interpreted from RMSE)
    plt.subplot(1, 2, 2)
    plt.plot(np.square(evals_result['val']['rmse']), label='Validation MSE')
    plt.title('Validation MSE Over Rounds')
    plt.xlabel('Rounds')
    plt.ylabel('MSE')
    plt.legend()

    plt.show()

    return model

def main_train():
    """
    Train XGBoost model and plot training progress.
    """
    X_train_split, X_val_split, y_train_split, y_val_split =\
     prepare_data_for_xgboost(X_train, y_train)

    model = train_xgb_model(X_train_split, X_val_split, y_train_split,
                            y_val_split, seed_value=42)

    # Save the trained model
    model.save_model('xgb_model.json')

if __name__ == "__main__":
    main_train()
```

## Evaluating the XGBoost Model

In this section, I focus on evaluating the performance of the trained XGBoost model. The evaluation process involves applying the model to test data and assessing its predictive accuracy using various metrics. This rigorous evaluation is crucial to understanding how well the model generalizes to new, unseen data.

### Workflow

1. **Ensuring Reproducibility**:

   - I begin by setting random seeds to ensure reproducibility in my results. This step is fundamental to obtaining consistent outcomes, which is critical for the validity of my evaluation process.

2. **Model Evaluation**:

   - The `evaluate_xgb_model()` function is central to this workflow, as it is responsible for assessing the model's performance. This function includes several important steps:

     - **Prediction**: I first convert the input features into the DMatrix format, which is optimized for XGBoost. Using this format, I generate predictions from the trained model.

     - **Reverse Scaling (if applicable)**:

      If a scaler was used during training to normalize the data, I reverse this scaling for both the predictions and the actual target values to ensure accurate evaluation metrics.

     - **Calculation of Metrics**:
     
      I compute key performance metrics, including Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), and the coefficient of determination (R²). These metrics provide a comprehensive overview of the model's predictive accuracy.

3. **Execution of the Main Evaluation Function**:

   - The `main_evaluate()` function orchestrates the evaluation process:
     - **Loading the Model**: I start by loading the pre-trained XGBoost model from a saved file.

     - **Evaluation on Test Data**:
     
     I apply the model to the test dataset, which has been kept separate from the training data to provide an unbiased assessment of the model's performance. The results, including MAE, RMSE, and R² for each ticker in the test dataset, are printed and stored.

     The metrics from both evaluations are compiled into a DataFrame for easy analysis and comparison. This DataFrame is saved in both CSV and Word formats for documentation and further reporting.

### Conclusion

By following this systematic approach, I thoroughly evaluate the XGBoost model's performance across multiple datasets. The use of detailed metrics such as MAE, RMSE, and R² allows me to critically assess the model's accuracy and generalizability. This evaluation is essential for ensuring that the model is not only effective on the training data but also performs well on unseen data, making it reliable for real-world applications.


```python
# Set random seeds for reproducibility
set_random_seeds()

def evaluate_xgb_model(model, X, y, scaler=None):
    """
    Evaluate XGBoost model performance.

    Args:
    - model: Trained XGBoost model.
    - X: Input features.
    - y: True target values.
    - scaler: Scaler to reverse scaling.

    Returns:
    - Dictionary with MAE, RMSE, R².
    """
    # Convert the data to DMatrix format
    dmatrix = xgb.DMatrix(X)

    # Predict using the XGBoost model
    predictions = model.predict(dmatrix)

    # If a scaler is provided, reverse the scaling of the predictions and targets
    if scaler is not None:
        y = scaler.inverse_transform(y.reshape(-1, 1)).flatten()
        predictions = scaler.inverse_transform(predictions.reshape(-1, 1)).flatten()

    # Calculate metrics
    mae = mean_absolute_error(y, predictions)
    rmse = np.sqrt(mean_squared_error(y, predictions))
    r2 = r2_score(y, predictions)

    return {"MAE": mae, "RMSE": rmse, "R²": r2}

def main_evaluate():
    """
    Main function to evaluate the XGBoost model and plot metrics.

    Returns:
    - None. Prints and plots metrics.
    """
    # Load the trained XGBoost model
    model = xgb.Booster()
    model.load_model('xgb_model.json')

    all_metrics = {}

    # Evaluate on test dataset
    print("=== Test Data ===")
    test_metrics = {}
    for ticker, data in preprocessed_data['test'].items():
        X_test_reshaped = data['X'].values
        y_test_reshaped = data['y'].values
        scaler = data.get('targ_scaler')
        metrics = evaluate_xgb_model(model, X_test_reshaped,
                                     y_test_reshaped, scaler)
        test_metrics[ticker] = metrics
        print(f"{ticker}: {metrics}")
    all_metrics["Test Data"] = test_metrics

    # Evaluate on extra test dataset
    print("\n=== Extra Test Data ===")
    extra_metrics = {}
    for ticker, data in preprocessed_data['extra'].items():
        X_extra_reshaped = data['X'].values
        y_extra_reshaped = data['y'].values
        scaler = data.get('targ_scaler')
        metrics = evaluate_xgb_model(model,
                                     X_extra_reshaped, y_extra_reshaped, scaler)
        extra_metrics[ticker] = metrics
        print(f"{ticker}: {metrics}")
    all_metrics["Extra Test Data"] = extra_metrics

    # Create DataFrame for results
    df = pd.DataFrame.from_dict(
        {(i, j): all_metrics[i][j]
         for i in all_metrics.keys()
         for j in all_metrics[i].keys()},
        orient='index')

    # Save and display table
    df.to_csv('xgb_evaluation_metrics.csv')
    save_as_word_table(df, "xgb_evaluation_metrics.docx")
    print(df)  # Display DataFrame directly

if __name__ == "__main__":
    main_evaluate()
```

```python
# Set random seeds for reproducibility
set_random_seeds()

def train_svm_model(X_train, y_train, X_val, y_val, kernel='rbf', C=1.0,
                    epsilon=0.1):
    """
    Train an SVM model using the provided training and validation data,
    display the training metrics, and plot the metrics.

    Args:
    - X_train: Training features.
    - y_train: Training targets.
    - X_val: Validation features.
    - y_val: Validation targets.
    - kernel: Kernel type to be used in the algorithm.
    - C: Regularization parameter.
    - epsilon: Epsilon in the epsilon-SVR model.

    Returns:
    - model: Trained SVM model.
    """
    model = SVR(kernel=kernel, C=C, epsilon=epsilon)
    model.fit(X_train, y_train)

    # Predictions on training and validation data
    train_predictions = model.predict(X_train)
    val_predictions = model.predict(X_val)

    # Calculate metrics
    train_mae = mean_absolute_error(y_train, train_predictions)
    train_mse = mean_squared_error(y_train, train_predictions)
    val_mae = mean_absolute_error(y_val, val_predictions)
    val_mse = mean_squared_error(y_val, val_predictions)

    # Display the training and validation metrics
    print("Training Metrics:")
    print(f"Training MAE: {train_mae:.4f}")
    print(f"Training MSE: {train_mse:.4f}")

    print("\nValidation Metrics:")
    print(f"Validation MAE: {val_mae:.4f}")
    print(f"Validation MSE: {val_mse:.4f}")

    # Plotting the training and validation metrics
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.bar(['Training MAE', 'Validation MAE'], [train_mae, val_mae])
    plt.title('MAE on Training and Validation Data')
    plt.ylabel('MAE')

    plt.subplot(1, 2, 2)
    plt.bar(['Training MSE', 'Validation MSE'], [train_mse, val_mse])
    plt.title('MSE on Training and Validation Data')
    plt.ylabel('MSE')

    # Explicitly show the plots
    plt.show()

    return model

def main_train_svm():
    """
    Main function to combine datasets, split the data, train the SVM model,
    and plot the training metrics.

    This function performs the following tasks:
    1. Combines the preprocessed datasets into a single training set.
    2. Splits the data into training and validation sets.
    3. Trains the SVM model using the training data.
    4. Plots the training and validation metrics.

    Returns:
    - model: Trained SVM model.
    """
    # Combine the datasets
    X_train, y_train = combine_datasets(preprocessed_data)

    # Split the data into training and validation sets
    X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
        X_train, y_train, test_size=0.2, random_state=42)

    # Train the SVM model and plot training metrics
    model = train_svm_model(X_train_split, y_train_split, X_val_split, y_val_split)

    return model

if __name__ == "__main__":
    model = main_train_svm()
```

```python
def evaluate_svm_model(model, X, y, scaler=None):
    """
    Evaluate SVM model performance.

    Args:
    - model: Trained SVM model.
    - X: Input features.
    - y: True target values.
    - scaler: Scaler to reverse scaling.

    Returns:
    - Dictionary with MAE, MSE, R².
    """
    # Predict using the SVM model
    predictions = model.predict(X)

    # If a scaler is provided, reverse the scaling of the predictions and targets
    if scaler is not None:
        y = scaler.inverse_transform(y.reshape(-1, 1)).flatten()
        predictions = scaler.inverse_transform(predictions.reshape(-1, 1)).flatten()

    # Calculate metrics
    mae = mean_absolute_error(y, predictions)
    mse = mean_squared_error(y, predictions)
    r2 = r2_score(y, predictions)

    return {"MAE": mae, "MSE": mse, "R²": r2}

def main_evaluate_svm():
    """
    Main function to evaluate the SVM model and plot metrics.

    Returns:
    - None. Prints and plots metrics.
    """
    all_metrics = {}

    # Evaluate on test dataset
    print("=== Test Data ===")
    test_metrics = {}
    for ticker, data in preprocessed_data['test'].items():
        X_test = data['X'].values
        y_test = data['y'].values
        scaler = data.get('targ_scaler')
        metrics = evaluate_svm_model(model, X_test, y_test, scaler)
        test_metrics[ticker] = metrics
        print(f"{ticker}: {metrics}")
    all_metrics["Test Data"] = test_metrics

    # Evaluate on extra test dataset
    print("\n=== Extra Test Data ===")
    extra_metrics = {}
    for ticker, data in preprocessed_data['extra'].items():
        X_extra = data['X'].values
        y_extra = data['y'].values
        scaler = data.get('targ_scaler')
        metrics = evaluate_svm_model(model, X_extra, y_extra, scaler)
        extra_metrics[ticker] = metrics
        print(f"{ticker}: {metrics}")
    all_metrics["Extra Test Data"] = extra_metrics

    # Create DataFrame for results
    df = pd.DataFrame.from_dict(
        {(i, j): all_metrics[i][j]
         for i in all_metrics.keys()
         for j in all_metrics[i].keys()},
        orient='index')

    # Save and display table
    df.to_csv('svm_evaluation_metrics.csv')
    save_as_word_table(df, "svm_evaluation_metrics.docx")
    print(df)  # Display DataFrame directly

if __name__ == "__main__":
    main_evaluate_svm()
```
